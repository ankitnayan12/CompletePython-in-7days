# Indexing with iloc
It is purely integer-location based indexing for selection by position

Allowed inputs are:

1) An integer, e.g. 5.

2) A list or array of integers, e.g. [4, 3, 0].

3) A slice object with ints, e.g. 1:7.

4) A boolean array.

5) A callable function with one argument (the calling Series, DataFrame or Panel) and that returns valid output for indexing (one of the above).

```python
# 1. Selecting a single row using iloc
```

```python
df_slice=df.iloc[0]    # It presents the 2nd row of the dataframe.
df_slice
```

1. Selecting row and column number using iloc

```python
df_slice1=df.iloc[1,2]  #It represents the value present in the ith row and ith column ie Algeria. You can cross-verify by checking the dataframe.
df_slice1
```

2. Selecting range of rows and a specified column

```python
df_slice2=df.iloc[1:5,2] # It represents the values of 1,2,3,4 rows and 2nd column.
df_slice2
```

3. Accessing multiple rows and multiple column values at a time.

```python
df_slice3=df.iloc[4:8,3:7]
df_slice3
```

4. Selecting all rows and specified columns using iloc

```python
df_slice4=df.iloc[:,5:10] #
df_slice4
```

5. Selecting all rows and columns starting from 0th position to specified column position

```python
df_slice5=df.iloc[:,:4]
df_slice5
```

6. Passing values to iloc() using lambda function.

```python
df_slice_func = df.iloc[lambda x: x.index % 2 == 0]   # We see all even indexes being selected.
df_slice_func.head()
```

# Combining all data frames into a single data using "pd.concat" function.

```python
rest_data = pd.concat([usercuisine,payment,profile,rest_accepts,rest_cuisine,rest_hours,rest_parking,geo_loc, rating_final])
rest_data.head()
```

We observe that entire data is now copied into a single dataframe. 
Concatenating technique is similar to Union used in set theory, that is the reason we observe null values in it.

Let us modify above "concat" function by including "ignore_index" parameter in it. This parameter is very beneficial in indexing as different dataframes shows it's own index while combining and can lead to a ambiguity. Inorder to remove ambiguity we can make use of ignore_index option.

```python
rest_data = pd.concat([usercuisine,payment,profile,rest_accepts,rest_cuisine,rest_hours,rest_parking,geo_loc,rating_final], ignore_index=True)
rest_data.head()
```

# Join DataFrames

We can also combine various dataframes using the "Join" operation based on some common column/key.
The logic behind these joins is similar to SQL where you join tables.

# Full-Outer Join

Full Outer Join is similar to concat() function discussed above, where it combines the results of both left and right outer joins. The joined dataframe contains all records from both dataframes and fills null values for missing matches on either side. We can perform full outer join by specifying the "how" argument as "outer" in the merge() function.

```python
rest_full_outer_join = pd.merge(rest_accepts, rest_cuisine, on='placeID', how='outer')
```

```python
rest_full_outer_join.head()
```

# Inner Join

Inner Join produces only set of records that matches in both dataframes. We have to pass "inner" in the "how" argument of the merge() function.

```python
rest_inner_join = pd.merge(rest_accepts, rest_cuisine, how='inner')
```

```python
rest_inner_join.head()
```

Thus, we can observe that through "Full Outer Join" records obtained is around 1920 rows and using "Inner Join" it is 1441 rows. Thus, 14.25% of data is reduced while using "Inner Join".

# Right Join

"Right Join" produces a complete set of records of dataframe in the right i.e "rest_cuisine", with the matching records in left dataframe "rest_accepts" . If there is no match, the left side will contain null. We need to pass "right" to the "how" argument of merge() function to do right join.

```python
rest_right_join=pd.merge(rest_accepts,rest_cuisine,how='right')
rest_right_join.head()
```

# Left Join

Left Join produces a complete set of records from DataFrame "rest_accepts" in the left with the matching records in the right dataframe "rest_cuisine". If there is no match, right side will contain null.Here, we need to pass "left" to "how" argument of the merge() function :

```python
rest_left_join = pd.merge(rest_accepts, rest_cuisine, on='placeID', how='left')
```

```python
rest_left_join
```

# Joining on Index

At times we need to perform the join on the indexes or the row labels. To do so, we have to specify the right_index and left_index as True:

```python
rest_index_df = pd.merge(rest_accepts, rest_cuisine, right_index=True, left_index=True)
```

```python
rest_index_df.head()
```

Thus on joining by index we get total of 916 records.

# Date and Time in Pandas

Pandas mainly combines the ease-of-use of "datetime" and "dateutil", vectorized interface of numpy.datetime64. From a group of these objects pandas can construct a DateTime Index that can be used to index data in a "Series" or "DataFrame".
