1.3 Colormaps

Colormaps map numerical values to a range of colors.
A simple way to create a custom linear colormap is to prepare a list of colors and allow Matplotlib to handle the transition.
The default colormap has been changed from 'jet', which spans the visible light spectrum from red to blue,to 'viridis', 
which is a perceptually uniform continuum from yellow to blue.

```python
import matplotlib.pyplot as plt
# Create a 30 random dots
np.random.seed(52)
x,y,c = zip(*np.random.rand(30,3))

# Create a linear colormap
cmap = matplotlib.colors.LinearSegmentedColormap.from_list("",["red","yellow","green"])

# Plotting the colors
plt.scatter(x,y,c=c,cmap=cmap)
plt.colorbar()
plt.show()
```

# Histogram

Histograms are useful in surveying the distribution of data. 
For example, we can plot data on a histogram when we want to see some age groups distributed in a population, 
light exposure in a photograph, or the amount of precipitation in each month in a city.

In Matplotlib, we call the plt.hist() function with a linear array. 
Matplotlib will automatically group them into 'bins' and plot out the frequencies for each bin in bars.

```python
# Example of plotting a randomly generated binomial distribution
np.random.seed(8)
x = np.random.binomial(100,0.5,size=10000)
plt.hist(x,bins=20) # plt.hist(x,20)
plt.show()
```

# Bar Plot
Bar plots are useful for comparing absolute levels of discrete data series. 
They are created by the function plt.bar(labels,heights) in Matplotlib.

```python
# Let's take an example of the market capitalization of today's much hyped cryptocurrencies.

# Data series consists of top 5 cryptocurrencies in terms of market capitalization
cc =['BTC','XRP','ETH','BCH','ADA']
cap = [282034,131378,107393,49999,26137]

# Plot the bar chart
plt.bar(cc,cap)
plt.title('Market capitalization of five top cryptocurrencies in Jan')
plt.xlabel('Crytocurrency')
plt.ylabel('Market capitalization (million USD)')
plt.show()
```

```python
import pandas as pd

# Create a dataframe
df=pd.DataFrame({'cc':cc,'cap':cap})

ax=df.plot(kind='bar')
ax=df[['cc','cap']].plot(kind='bar') # in case df has multiple columns 
ax.set_xticklabels(df['cc'])

plt.title('Market capitalization of five top cryptocurrencies in Jan')
plt.xlabel('Crytocurrency')
plt.ylabel('Market capitalization (million USD)')
plt.show()
```

```python
# To draw a multibar plot to show the performance of an imaginary medicine to treat cancer, by 
# comparing the level of an cancer level before and after treatment of a medicine and placebo 
# as control 

# Prepare the data series 
import pandas as pd 
labels_medicine = ['Medicine(Before)','Medicine(After)'] 
labels_placebo = ['Placebo(Before)', 'Medicine(After)']
medicine = [2.88,1.42]
placebo = [2.72,2.68]
yerr_medicine = [0.12,0.08]
yerr_placebo = [0.24,0.13]

df = pd.DataFrame([medicine,placebo])
df.columns = ['Before','After']
df.index = ['Medicine','Placebo']

# Plot the bar chart with error bars
df.plot(kind='bar',width=0.4,color=['midnightblue','cornflowerblue'],yerr=[yerr_medicine,yerr_placebo])

plt.title('Effect of Medicines')
plt.xlabel('Condition')
plt.ylabel('[hsCRP] (mg/L)')
plt.xticks(rotation=0) # to keep the xtick labels horizontal
plt.legend(loc=(0.4,0.8))

plt.show()
```

# Box Plot
Box Plots are most suited for displaying the distribution of a variable across multiple groups. 
The bottom and top of the boxes indicate the lower and upper quartiles, respectively, and the line inside the box is for the median.
Vertical lines extending from the boxes("whiskers") show the range of the data(this is 1.5x past the upper and lower quartiles in matplotlib).

Box Plots can be thought of as an hybrid between histograms and bar plots.
They surface much of the same information as bar plots, but they also expose the variation in the data.

# Pie Chart
Pie chart is a circular representation of component ratios.
The angle and hence the arc length of each sector ratio(called 'wedges') presents the proportion that each component accounts for , relative to the whole.
Matplotlib uses plt.pie() function to draw pie charts. 
We can label each sector with 'labels' as well as percentage with 'autopct' automatically.

```python
# Example of web server using in first week of January-2017
x=[0.31,0.3,0.14,0.1,0.15]
labels=['nginx','Apache','IIS','Varnish','Others']
plt.pie(x,labels=labels,autopct='%1.1f%%')
plt.title('Web Server Usage Statistics')
plt.show()
```

# Drawing Subplots
When you have multiple plots it is necessary to organize them into panels in the same figure, such as when illustrating different aspects of the same dataset.
Matplotlib provides a few ways to create figures with multiple subplots.

```python
# Initiates a figure area for plotting
fig = plt.figure()
# Initiates six subplot axes
ax1 = plt.subplot(231)
ax2 = plt.subplot(232)
ax3 = plt.subplot(233)
ax4 = plt.subplot(234)
ax5 = plt.subplot(235)
ax6 = plt.subplot(236)

# Print the type of ax1
print(type(ax1))

# Label each subplot with corresponding identities
ax1.text(0.3,0.5,'231',fontsize=18)
ax2.text(0.3,0.5,'232',fontsize=18)
ax3.text(0.3,0.5,'233',fontsize=18)
ax4.text(0.3,0.5,'234',fontsize=18)
ax5.text(0.3,0.5,'235',fontsize=18)
ax6.text(0.3,0.5,'236',fontsize=18)

plt.show()
```
